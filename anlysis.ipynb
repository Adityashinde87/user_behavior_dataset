```python
# Q1 Which Gender use more smart phone ?

ax = sns.countplot(x= 'Gender' , data = df)

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
# Q2 Which oprating system use most ?

ax = sns.countplot(x='Operating System', data = df)

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
# Q3 Find each smartphone brand device Using Avrage time spend in day ?

using_time = df.groupby(['Device Model'], as_index= False)['App Usage Time (min/day)'].mean().sort_values(by='App Usage Time (min/day)', ascending=False)

ax = sns.set(rc={'figure.figsize':(15,5)})
ax = sns.barplot(data= using_time, x ='Device Model', y = 'App Usage Time (min/day)')

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
# Q4 Find each smartphone brand device Using Avrage battery drain in day ?

user_battery = df.groupby(['Device Model'] , as_index=False)['Battery Drain (mAh/day)'].mean().sort_values(by = 'Battery Drain (mAh/day)', ascending=False)

ax = sns.set(rc={'figure.figsize':(10,5)})
ax = sns.barplot(x= 'Device Model' , y= 'Battery Drain (mAh/day)', data = user_battery)

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
# Q5 Which smartphone brand device installed more applications from user ?

user_app = df.groupby(['Device Model'] , as_index=False)['Number of Apps Installed'].mean().sort_values(by = 'Number of Apps Installed', ascending=False)

ax = sns.set(rc={'figure.figsize':(10,5)})
ax = sns.barplot(x = 'Device Model' , y = 'Number of Apps Installed' , data = user_app )

for bar in ax.containers:
    ax.bar_label(bar)
    
```

```python
# Q6 Which smartphone brand use user gender wise count ?

ax = sns.countplot(x =  'Device Model' , hue = 'Gender' , data = df)

for bar in ax.containers:
    ax.bar_label(bar)
    
```

```python
# Q7 Minimum age of user using smarthphone brand device wise ?

user_age = df.groupby('Device Model', as_index=False)['Age'].min().sort_values(by = 'Age' ,ascending = False)

ax = sns.set(rc={'figure.figsize':(10,5)})
ax = sns.barplot(x = 'Device Model' , y = 'Age' , data = user_age)

for bar in ax.containers:
    ax.bar_label(bar)
              
```

```python
# Q8 Find each smartphone brand device Avrage of user behavior class ?

user_class = df.groupby('Device Model', as_index=False)['User Behavior Class'].mean().sort_values(by = 'User Behavior Class' ,ascending = False)

ax = sns.set(rc={'figure.figsize':(10,5)})
ax = sns.barplot(x = 'Device Model' , y = 'User Behavior Class' , data = user_class)

for bar in ax.containers:
    ax.bar_label(bar)
```
